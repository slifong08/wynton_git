## load labels

```python
def processLabel(label_file, include_activity, concat):
    """
    make pd dataframe w meta info from label file
    
    input
        label_file (str) - sei label file
        include_activity(bool) - flag to add info if label file has activty percent rank annotation
        concat (bool) - flag if construct was concat
        
    method
        1. read sei label file
        2. add coordinate annotations
        3. deal with endog annotation error for concat
        4. annotate if activity
        5. annotate if concat
        
    return 
        lab (dataframe) - dataframe of label meta data
    """
    
   
    #1 read label file as pd dataframe
    lab = pd.read_csv(label_file,sep ='\t')
    lab = lab.drop(columns = ["index"])  # redundant index column
    

    # str splitting
    #2 coordinates
    lab["coor"]= lab["name"].apply(lambda x: x.split("_")[0])

    #3 if endog annot didnt work
    if lab["name"].iloc[0] == "None":
        coor = lab["coor"].iloc[1]
        lab["name"].iloc[0] = f"{coor}_-1_endog"
        lab["coor"].iloc[0] = coor

    #4 if activtiy scores in dataframe
    if include_activity is True:
        lab["avg_pct"]= lab["name"].apply(lambda x: x.split("_")[1]).map(float)#.round(1) *10
        lab["bin"]= lab["name"].apply(lambda x: x.split("_")[1]).map(float).round(1) *10

    #5 concat number
    if concat is True:
        lab["id"] = lab["name"].apply(lambda x: x.split("_")[1])
        lab["type"] = lab["name"].apply(lambda x: x.split("_")[2])


    return lab
```

```python
def returnSequenceClassLabels():
    file="/wynton/home/ahituv/fongsl/bin/sei-framework/sequence_class_labels.csv"
    lab = pd.read_csv(file)
    
    return lab
```

```python
"""
# get sequence class labels. See Methods section of Chen 2022 for interpretation of these PC labels.
# apparently labels >40 are low active/heterochromatin. 
#Make up <2% of the genome. But 2% of the genome can still be significant.
"""

seqClass = returnSequenceClassLabels()

seqClass.head()
```

```python
ACTIVITY = False
# open npy data
data = np.load(
                os.path.join(PATH,
                 f"{HANDLE}.raw_sequence_class_scores.npy"),
                allow_pickle=True
                            )
labels=os.path.join(PATH,
                 f"{HANDLE}_row_labels.txt")

# turn into pd dataframe
df=pd.DataFrame(data)

# rename columns
df.columns=list(seqClass["#Sequence class label"])[:-1]

# process labels file
lab=processLabel(labels, ACTIVITY, True)

if ACTIVITY is True:
    act = pd.read_csv(ACT, sep='\t', usecols=[ "avg_activity", "coor.hg38",])
    act.columns = ["avg_activity","coor"]

    lab=pd.merge(lab, act, how="left")

# add labels and data together
df=pd.merge(lab, df, left_index=True, right_index=True)
```

## LASSO 

```python
lasso_model = Lasso(alpha=0.4, fit_intercept=False, max_iter=100)

lasso_model.fit(X, y["avg_activity"])

lasso_model.coef_, lasso_model.sparse_coef_, lasso_model.n_iter_, 
```

```python
lasso_cols = []
for i, n in enumerate(lasso_model.coef_):
    if abs(n)>0:
        print(datacols[i])
        lasso_cols.append(datacols[i])

        print(i, n)
```
