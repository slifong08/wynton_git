# analyze data

## load labels

```python
def processLabel(label_file):
    
   # read label file as pd dataframe
    lab = pd.read_csv(label_file,sep ='\t')
    lab = lab.drop(columns = ["index"])  # redundant index column


    # str splitting
    # coordinates
    lab["coor"]= lab["name"].apply(lambda x: x.split("_")[0])

    # the concat number
    lab["avg_pct"]= lab["name"].apply(lambda x: x.split("_")[1]).map(float)#.round(1) *10
    lab["bin"]= lab["name"].apply(lambda x: x.split("_")[1]).map(float).round(1) *10

    return lab
```

## sequence class labels

```python
def returnSequenceClassLabels():
    file="/wynton/home/ahituv/fongsl/bin/sei-framework/sequence_class_labels.csv"
    lab = pd.read_csv(file)
    
    return lab
```

```python
"""
# get sequence class labels. See Methods section of Chen 2022 for interpretation of these PC labels.
# apparently labels >40 are low active/heterochromatin. 
#Make up <2% of the genome. But 2% of the genome can still be significant.
"""

seqClass = returnSequenceClassLabels()

seqClass
```

# Modeling sei sequence class prediction performances

## continuous activity data x sequence class 
- train test split

```python
# predict activity from data columns
X, y = df[datacols].fillna(0), df["avg_activity"]

# split into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## lasso regression - alpha gridCV

```python
# select alpha for lasso regression? 
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
lasso_alphas = np.linspace(0, 100, 10)
grid = dict()
grid['alpha'] = lasso_alphas

sc = StandardScaler()
X_train_s = sc.fit_transform(X_train)
lasso=Lasso()
search = GridSearchCV(lasso, grid, cv=cv).fit(X_train_s, y_train)

cvres = search.cv_results_

print("best params, score", search.best_params_, search.best_score_)
```

```python
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

## linear regression 
    - because alpha is zero from gridCV

```python
# fit the pipeline
pipe = Pipeline([('scaler', StandardScaler()), ('lr', LinearRegression())])
pipe.fit(X_train, y_train)
```

```python
ypred = pipe.predict(X_test)

# root mean square error
lr_mse = mean_squared_error(y_test, ypred)
lr_rmse = np.sqrt(lr_mse)
lr_rmse
```

## randomly sample

```python
def getSampleCoord(activity_label, df):
    
    coors = []
    
    # get the coordinates linked to desired bin_id
    coors.extend(list((df.loc[df["act_label"]==activity_label, "coor"])))
    
    print(len(coors))
    # randomly sample 50
    random50 = np.random.choice(coors, size=50, replace=False)
    
    # return list of coordinates
    return coors, random50
```

```python
coors_active, random50active = getSampleCoord(1, logreg)
coors_inactive, random50inactive = getSampleCoord(0, logreg)
```

# OLS 

```python
# OLS on active, inactive labels

X, y = logreg[datacols].fillna(0), logreg[annot_cols]

# scale data
ss = StandardScaler()
Xt = ss.fit_transform(X)
Xtc = sm.add_constant(Xt)

mod = sm.OLS(y["avg_activity"], Xtc)

res = mod.fit()

print(res.summary())
```

```python
from statsmodels.stats.multitest import fdrcorrection
fdrbool, fdrp = fdrcorrection(res.pvalues)

sig_cols =[]
for n, v in enumerate(fdrbool):
    
    if bool(v) == True:
        sig_cols.append(n)
sig_cols

sig_OLS_cols=[]
for s in sig_cols[1:]:
    print(datacols[s-1])
    sig_OLS_cols.append(datacols[s-1])
```

```python
row_colors_, lut_ = getColorMap("act_label", logreg, cmap_)
cm_ols = sns.clustermap(logreg[sig_OLS_cols].drop_duplicates(), 
                   row_colors=row_colors_,
                    square=True,
                   figsize=(6,10),
                    z_score =1,

                   )
from matplotlib.patches import Patch
plt.setp(cm_ols.ax_heatmap.xaxis.get_majorticklabels(), fontsize=10)
plt.setp(cm_ols.ax_heatmap.yaxis.get_majorticklabels(), fontsize=6)

handles = [Patch(facecolor=lut_[name]) for name in lut_]
plt.legend(handles, lut_, title='activity_label',
           bbox_to_anchor=(1, 1), bbox_transform=plt.gcf().transFigure, loc='upper left')
plt.savefig(os.path.join(RE, "sei_pred_seqclass_minmax.n150.cluster.standard.OLS.pdf"), bbox_inches="tight")
```
