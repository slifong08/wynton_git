# functions 

## processing functions

```python
def readHdF5(filename, path):
    """
    read hdf5 file, dataset
    
    input
        filename (str) - name of hdf5 results
        path (str) - abs path to directory
        
    require 
        h5py
        
    method 
        1. assemble file
        2. read file
        3. get dataset

    return
        file, data
    
    """
    #1
    file = os.path.join(path, filename) 
    
    #2
    f = h5py.File(file, 'r')
    
    #3
    dset = f['data']

    return f, dset

def getIndexNames():
    
    """
    return list of 21097 index names corresponding to sei prediction features
    
    input
        none
    method 
        1. make a list for the index names
        2. open the index names file 
        3. append name to index name list
    
    return
        idxnames (list) - list of index names
    """
    
    #1
    idxnames = []
    
    #2
    IDXNAMES = "/wynton/home/ahituv/fongsl/bin/sei-framework/seq_prediction_columns.txt"
    with open(IDXNAMES, "r") as reader:
        for line in reader:
            #3
            idxnames.append(line.strip("\n"))
            
    return idxnames

def arrayToDF(dset):
    
    df = pd.DataFrame(np.vstack(dset))
    print(df.shape)
    df.columns = getIndexNames() # function get column names
    
    return df.reset_index()

def getRowNames(row_textfile, path):
    """
    return list corresponding to sample names
    """
    
     #1
    rownames = {}
    file = os.path.join(path, row_textfile) 
    
    #2
    with open(file, "r") as reader:
        for line in reader:
            
            #3
            idx, name = line.strip("\n").split("\t")
            if idx !="index":
                rownames[idx]=name
    df = pd.DataFrame(rownames.items(), columns= ["index_col", "id"])
    df["index_col"]=df["index_col"].astype(int)
    return df
```

## analysis functions

```python
def computeFeatureMean(feature, df):
    
    # pivot dataframe related to feature
    dif = pd.pivot(df.loc[df["feature"]==feature], index="pair_id", columns="track",
                                 values="pred_prob").reset_index()
    
    # calculate the mean score
    dif["mean"] = dif[dif.columns[1:]].mean(axis=1)
    
    # add feature as a column
    dif["feature"] = feature
    
    # calculate distance from kmer to each mutation
    dif["kmer_dif"] = dif["mean"] - dif.loc[dif["pair_id"].str.contains('kmer'), "mean"].iloc[0]
    
    # rank the means. 
    dif["mean_rank"]=dif["mean"].rank()
    
    dif["mean_z"]=stats.zscore(dif["mean"])
    
    return dif[["pair_id", "mean", "feature", "kmer_dif","mean_rank"]]
```

```python
def getPredValues(df, kmer, null):
    """
    use w/ dif dataframe only (feature-specific dataframe w/ means calculated)
    """
    kmer_pred = dif.loc[dif["pair_id"] == kmer]["mean"].iloc[0]
    null_pred = dif.loc[dif["pair_id"] == null]["mean"].iloc[0]
    ctrl_pred = dif.loc[~dif["pair_id"].isin([kmer, null]),
                        "mean"]

    ctrl_pred_mean = ctrl_pred.mean()
    ctrl_pred_std = ctrl_pred.std()
    
    return [kmer_pred, null_pred, ctrl_pred_mean, ctrl_pred_std]
```

```python
def makeLongForm(celldf):

    # melt dataframe into long form
    test = pd.melt(cdf, id_vars=["nullomer", "coor", "pair_id", "id", "pos"],
                           var_name="track", value_name="pred_prob")
    
    # annotate cl, feature, and dataset ids
    test["cl"] = test["track"].apply(lambda x:x.split("|")[0])
    test["feature"] = test["track"].apply(lambda x:x.split("|")[1])
    test["dataset"] = test["track"].apply(lambda x:x.split("|")[2])
    
    return test
```

```python
def getKmerNullIds(longdf):
    # kmer_id
    kmer = longdf.loc[longdf['pair_id'].str.contains(
                    "kmer"), "pair_id"].iloc[0]
                
    # nullomer_id
    null = longdf.loc[longdf['pair_id'].str.contains(
                    "null"), "pair_id"].iloc[0]

    return kmer, null
```

```python
def plotHist(dif, null_pred, kmer_pred, coor, feature):
        fig, ax = plt.subplots(figsize=(4, 4))

        # plot distribution of prediction differences
        sns.histplot(dif["mean"])

        # plot null difference
        ax.axvline(null_pred, ls="--", c="r")
        ax.axvline(kmer_pred, ls="--", c="b")

        # name the plot
        ax.set(title=feature,
               xlabel=f"pred - {COOR}"
               )
        plt.show()
```

```python
def infoVector(coor, tcoor, kmer, null, feature, feature_summary_stats):

    # make info vector to write
    kmer_pred, null_pred, ctrl_pred_mean, ctrl_pred_std = feature_summary_stats

    info_vector = [
        coor, tcoor,
        kmer, null,  # pair ids
        feature,  # prediction track
        str(kmer_pred),
        str(null_pred),
        str(ctrl_pred_mean),
        str(ctrl_pred_std),
    ]

    columns = ["#coord", "tsvCoor", "kmer", "null",
               "track", "predKmer", "predNull",
               "predCtrl_mean", 'predCtrl_std',
               ]

    return info_vector, columns
```

# load data

## load hdf5

```python
n = len(glob.glob(QUERY_F))
n
```

```python
collection = {}
for F in (glob.glob(QUERY_F)):

    F_LABELS = F.split("_predictions.h5")[0] + "_row_labels.txt"
    f, dset = readHdF5(F,PATH)

    df = arrayToDF(dset)

    ## get row names

    names = getRowNames(F_LABELS, PATH)

    # add to dataframe
    df = pd.merge(names, df, how="left", left_on ="index_col", right_on = "index")

    # drop the index_col name (redundant)
    df = df.drop(columns=["index_col", "index"])

    # format dataframe

    df["coor"] = df["id"].apply(lambda x: x.split("_")[0])
    df["pos"] = df["id"].apply(lambda x: x.split(".")[-1] if "." in x else None)
    df["pair_id"] = df["id"].apply(lambda x: (x.split("_")[1]).split(".")[0])

    # add kmer info
    df["nullomer"] = False 
    df.loc[df["id"].str.contains("null"), "nullomer" ] = True



    df = df.loc[df.pos.astype(str) != "None"]  # remove all the extra repeats of the same kmer control data
    collection[n]= df
```

```python
# combine
df = pd.concat(collection.values())
print(df.shape)
df.head(33)
```

# analyze one locus

```python
out = os.path.join(PATH, "test15_HepG2.K562.Pred.tsv")
with open(out, "w") as writer:

    v = 0
    for COOR in test_coor:

        # COOR = "chr1:10248183-10248383"
        TCOOR = '.'.join(COOR.split(":"))  # stupid formatting

        coordf = df.loc[(df["coor"] == COOR)].copy()

        # Test nullomer prediction differences on locus at at time.

        # Separate by prediction positions
        # can be multiple places where kmer/nullomer occurs in sequence.
        for position in set(coordf["pos"]):
            cdf = coordf.loc[coordf["pos"] == position].copy()

            # melt dataframe into long form
            longdf = makeLongForm(cdf)

            # keep only datasets w/ controls
            if len(set(longdf.nullomer)) > 1:

                # get the kmer, nullomer id
                kmer, null = getKmerNullIds(longdf)

                # take mean score for every feature dataset, row-wise
                for feature in set(longdf["feature"]):

                    # compute the means, change in means from kmer
                    dif = computeFeatureMean(feature, test)

                    # get summary stats about the feature means, ctrl variation
                    feature_summary_stats = getPredValues(dif, kmer, null)

                    # unpack summary stats variables
                    kmer_pred, null_pred, ctrl_pred_mean, ctrl_pred_std = feature_summary_stats

                    # if nullomer z score is 2 sd away.
                    if abs(ctrl_pred_std) > 0.05 and null_pred > 0.1:
                        

                        if v < 5:
                            print(feature, COOR, abs(null_pred - ctrl_pred_mean))
                            
                            # plot
                            plotHist(dif, null_pred, kmer_pred, COOR, feature)

                        # write data to file
                        writeinfo, columns = infoVector(COOR, TCOOR, kmer, null,
                                                        feature, feature_summary_stats)
                        if v == 0:
                            writer.write("\t".join(columns)+"\n")

                        writer.write("\t".join(writeinfo)+"\n")

                        v += 1
    writer.close()
```
