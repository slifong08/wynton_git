```python
def prettifySeq(original, mut):
    """
    prettify sequence. 
        All matching bases will be written as "."
        All non matching bases will be written w mutated base identity. 
    """
    prettyseq = ""
    for o, m in zip(original, mut):
        if o != m:
            prettyseq += m
        else:
            prettyseq += "."

    return prettyseq
```

# get matched controls 2 bp away

```python
seqs = zip(ndf["kmer"], ndf["matched_seq"])

control_dict = {}
kmer_resurface, nulls = set(), set()
for original_seq, null_seq in seqs:
 
    control_set = buildOtherSeqMuts(original_seq, null_seq)

    kmer_controls = control_set.difference(null_ref)  # subtract any nullomers created from the mutations
    
    control_dict[null_seq]=kmer_controls
    kmer_resurface.update(kmer_controls)
    nulls.add(null_seq)
```

```python
len(control_dict.keys()), len(kmer_resurface)
```

```python
# make dataframe of nullomers and matched controls. 
ctrl = pd.DataFrame(control_dict.items(), columns = ["null", "ctrl_kmers"])
```

# control for GC content

```python
e = ctrl.explode("ctrl_kmers")
e["ctrl_gc"] = e['ctrl_kmers'].apply(lambda x: gc_fraction(x))
e["null_gc"] = e['null'].apply(lambda x: gc_fraction(x))
```

```python
e = e.loc[(e["ctrl_gc"] <= e["null_gc"].max()) &
        (e["ctrl_gc"] >= e["null_gc"].min()) ]
```

```python
e.describe()
```

```python
e.groupby(["ctrl_kmers", "ctrl_gc"]).count().sort_values(by = "null", ascending =False).reset_index()
```

```python
popular_list = list(e.groupby("ctrl_kmers").count().sort_values(by = "null", ascending =False).reset_index()["ctrl_kmers"])
```

```python
n = nulls.copy()

for num, kmer in enumerate(popular_list):
    remove_nulls = set(e.loc[e["ctrl_kmers"]==kmer, "null"])
    for i in remove_nulls:
        if i in n:
            n.remove(i)
            
            print(len(n))
    if len(n)==0:
        print("top", num)
        break
    else:
        continue
```
