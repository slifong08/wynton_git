## Functions

```python
def returnSequenceClassLabels():
    file = "/wynton/home/ahituv/fongsl/bin/sei-framework/sequence_class_labels.csv"
    lab = pd.read_csv(file)

    return lab


def processLabel(label_file, label_index):
    """
    input 
        label_file  (str) - path with the labels for the sequences run through SEI
        label_index (str) - path with the labels for the sequences run through SEI and their original labels

    Method
        1. opens sequence label file, label_index as pd dataframe
        2. drop the index column
        3.  str split to get sequence name, null.id, cell line, the scaffold.id, 
            and clean name to match MPRA dataframe
            
    return 
        lab (pd dataframe) - dataframe with label annotations
    """

    # read label file as pd dataframe
    lab = pd.read_csv(label_file, sep='\t', skiprows=1, names=["index", "seq.id"])


    labi = pd.read_csv(LABEL_INDEX, sep='\t', header=None, names=["seq.id", "seq.name"])

    lab = pd.merge(lab, labi)

    lab = lab.drop(columns=["index"])  # redundant index column

    # str splitting - replace any underscore w pipe
    lab["seq.name"] = lab["seq.name"].apply(lambda x: "|".join(x.split("_")))

    # coordinates
    lab["null.id"] = lab["seq.name"].apply(lambda x: x.split("|")[0])

    # the concat number

    lab["CL"] = lab["seq.name"].apply(lambda x: x.split("|")[1])

    lab["scaffold.id"] = lab["seq.name"].apply(lambda x: x.split("|")[2])

    lab["clean_name"] = lab["CL"] + "_"  + lab["scaffold.id"]  # for joining w/ MPRA activity dataframe

    return lab


def processLibrary(library_file):
    """
    load and format the library_file dataframe.

    input 
        library_file (str) - path to file

    method 
        1. load library file as pd dataframe
        2. remove any sequences without hg38 coordinates (synthetic/scrambled controls)
        3. compute GC content
        4. change coordinate datatype to int
        5. make coordinate column
    return
        lib (pd dataframe) - dataframe w/ endogenous MPRA library information

    """
    lib_ = pd.read_csv(library_file, header=1)
    print(lib_.shape)

    # remove names that do not have hg38 coordinates (n=670)
    not_hg38 = lib_.loc[(lib_["chr.hg38"].isna()), "name"].copy()

    # remove names w/o coordinate information (to make coordinate label)
    lib = lib_.loc[~lib_["name"].isin(not_hg38)].copy()

    # gc faction
    lib["gc"] = lib[
        "230nt sequence (15nt 5' adaptor - 200nt element - 15nt 3' adaptor)"]\
        .apply(lambda x: gc_fraction(x[15:-15]))

    # annotate the coordinates of the MPRA sequences
    lib[["start.hg38", "stop.hg38"]] = lib[
        ["start.hg38", "stop.hg38"]].astype(int)

    # make coordinate column
    lib["coor.hg38"] = "chr" + lib["chr.hg38"] + ":" + \
        lib["start.hg38"].map(str) + "-" + lib["stop.hg38"].map(str)

    
    print(lib.shape)

    return lib
```

```python
def clusterMapCL(df, plot_columns, **kwargs):
    color_col = kwargs.get('color_col', None)
    color_col2 = kwargs.get('color_col2', None)
    row_cluster = kwargs.get('row_cluster', True)
    col_cluster = kwargs.get('col_cluster', True)

    # color activity bin
    if color_col is not None and color_col == "avg_activity":
        df["act_bin"] = 0
        df.loc[df[color_col] > 1, "act_bin"] = 1

        # map colors

        lut = dict(zip(df["act_bin"].unique(),

                       ["#c1f80a", "#fdaa48"]
                       ))
        row_colors1 = df["act_bin"].map(lut)
        print(lut)

    # color cell lines
    if color_col2 is not None and color_col2 == "CL":
        cmap = matplotlib.colormaps["tab10"]

        rgba = cmap(0.5)[:3]

        df["CL_bin"] = 0
        lut2 = dict(zip(df[color_col2].unique(),
                        [cmap(0.4), cmap(0.2), cmap(0.3)]  # list of colors
                        ))

        row_colors2 = df[color_col2].map(lut2)

        print(lut2)
        row_colors = pd.concat([row_colors2, row_colors1], axis=1)

        # clustermap
        g = sns.clustermap(df[plot_columns],
                           cmap="bwr",
                           center=0,
                           row_colors=row_colors,
                           vmin=-50,
                           vmax=50,
                           col_cluster=col_cluster,
                           row_cluster=row_cluster,
                           figsize=(15, 15),
                           cbar_kws={'label': 'sei pred'}
                           )

        # for legend plotting
        handles = [Patch(facecolor=lut[name]) for name in lut]
        handles2 = [Patch(facecolor=lut2[name]) for name in lut2]

        # plot CL legende
        plt.legend(handles2, lut2, title='CL',
                   bbox_to_anchor=(0.8, 1),
                   bbox_transform=plt.gcf().transFigure,
                   loc='upper left')

        # plot the activity legend
        for label in [0, 1]:
            g.ax_row_dendrogram.bar(
                0, 0, color=lut[label], label=label, linewidth=0)
            l3 = g.ax_row_dendrogram.legend(title='activity',
                                            loc='upper right', ncol=1,
                                            bbox_to_anchor=(5.9, 1.15))

    else:
        g = sns.clustermap(df[plot_columns],
                           cmap="bwr",
                           center=0,
                           vmin=-50,
                           vmax=50,
                           col_cluster=col_cluster,
                           row_cluster=row_cluster,
                           figsize=(15, 15),
                           cbar_kws={'label': 'raw sei pred'}
                           )
    return g
```

```python
def reindexDfRows(df, col_name, row_order_list):
    """re-index rows for a df column based on an ordered list"""

    # set index
    df = df.set_index(col_name)
    
    # returm sorted dataframe w/ index reset
    return df.loc[row_order_list].reset_index()  

def reindexDfCols(df, col_list, **kwargs):
    """re-order df columns by list"""
    
    # someitmes, you want to freeze some of the columns
    freeze_cols = kwargs.get("freeze_cols", None)
    
    if freeze_cols is None:
        reorderd = df[col_list]
    else:
        print("freeze", list(df.columns[:freeze_cols]))
        new_col_list = list(df.columns[:freeze_cols]) + list(col_list)
        reorderd = df[new_col_list]
    
   
    
    return reorderd
```
