## Functions

```python
def returnSequenceClassLabels():
    file = "/wynton/home/ahituv/fongsl/bin/sei-framework/sequence_class_labels.csv"
    lab = pd.read_csv(file)

    return lab

def getSeparateDf(df):
    """
    separate any dataframe into endog and nullomer specific dataframes

    input 
        df (pd dataframe) - a dataframe of the predictions. Must have a "null.id" column

    methods
        1. separate the endogenous with str search - where null.id column contains "endog" 
        2. separate the nullomers with str search - where null.id column contains "firstorder"

    return 
        endog_df (pd.df) - dataframe of endoegnous sequences only
        null_df (pd.df) - dataframe of nullomer sequences only
        scram_df (pd.df) - dataframe of scrambled sequences

    """

    # 1 endog dataframe
    endog_df = df.loc[df["null.id"]=="endog"].copy()

    # 2 null dataframe
    null_df = df.loc[df["seq.name"].str.contains("firstorder")].copy()

    # 3 scramble dataframe
    scram_df = df.loc[df["seq.name"].str.contains("scramble")].copy()

    #print("TEST", endog_df.shape, null_df.shape, scram_df.shape)
    
    return endog_df, null_df, scram_df


def OneSeqManyNulls(seq_coor, no_concatdf, datacols, annot_cols):
    """
    focus on 1 enhancer sequence, and effects of many nullomers

    input
        seq_coor (str) - coordinate of enhancer sequence
        no_concatdf (pd df) - dataframe of sei predictions for 40 sequence classes, 
            across enhancer sequences and nullomer inserts
        datacols (list) - columns with the 40 sequence classes
        annot_cols (list) - annotation columns for sequences tested in sei
        
    method
        1. subset dataframe for single seq_coor
        2. split scaled dataframe into endogenous and nullomer sequences
        3. subtract the endogenous activity vector from the nullomer sequence matrix using sequeeze function
        4. do the same for the scrambled sequences
    return
        test_endog (pd df)- standard scaled enhancer sequence prediction values (vector/ 1 row)
        test_null (pd df) - standard scaled nullomer prediction values
        test_sub (pd df) - difference between endogenous and nullomer predictions in scaffold
    """
    # pick one coordinate.
    # coor = "chr10:102347343-102347543"

    # 1 get the coordinate endogenous predictions
    onecoor = df.loc[df["coor"] == seq_coor].copy()
    activity = onecoor["avg_activity"].iloc[0]

    # 2 split dataframes into endog + null
    #test_endog, test_null = getSeparateDf(onecoor)
    endog_df, null_df, scram_df = getSeparateDf(onecoor)
    

    # 3 subtract theh endogenous from the nullomers to get vector of nullomers
    obs_sub = null_df[datacols].sub(endog_df[datacols].squeeze(), axis=1)

    obs_sub = pd.merge(obs_sub, null_df[annot_cols],
                        left_index=True, right_index=True)


    # subtract the scramble from the nullomers to get vector of nullomers
    exp_sub = scram_df[datacols].sub(endog_df[datacols].squeeze(), axis=1)
    exp_sub = pd.merge(exp_sub, scram_df[annot_cols],
                        left_index=True, right_index=True)

    
    return endog_df, null_df, obs_sub, onecoor, exp_sub


def seqClassLookup(annot_list):
    print(seqClass.loc[seqClass[seqClass.columns[0]].isin(annot_list)])


def getColorMap(col, df, cmap):
    
    lut = dict(zip(df[col].unique(), cmap))
    print(lut)

    row_colors = df[col].map(lut)

    return row_colors, lut

def plotClusterMap(table, outhandle, seq_coor, **kwargs):
    color_col = kwargs.get('color_col', None)
    
    
    if color_col is not None:
        table = table.set_index("null.id")
        lut = dict(zip(table[color_col].unique(), ["white", "grey"]))
        print(lut)

        row_colors = table[color_col].map(lut)
        
        sns.clustermap(table[table.columns[:-7]], 
                       cmap="bwr", center=0,
                       vmin=-10, vmax=10,
                       figsize=(15, 15),
                       row_colors = row_colors)
        
        # plot legend
        handles = [Patch(facecolor=lut[name]) for name in lut]
        plt.legend(handles, lut, title='Obs/Exp',
           bbox_to_anchor=(0.8, 1), bbox_transform=plt.gcf().transFigure, loc='upper left')
    else:
        sns.clustermap(table, cmap="bwr", center=0,
                   vmin=-10, vmax=10,
                       figsize=(15, 15))
        
    plt.savefig(os.path.join(RE, f"{outhandle}.{seq_coor}.pdf"))
    plt.show()
```

## clustermap CL function

```python
def clusterMapCL(df, plot_columns, **kwargs):
    color_col = kwargs.get('color_col', None)
    color_col2 = kwargs.get('color_col2', None)
    row_cluster = kwargs.get('row_cluster', True)
    col_cluster = kwargs.get('col_cluster', True)

    # color activity bin
    if color_col is not None and color_col == "avg_activity":
        df["act_bin"] = 0
        df.loc[df[color_col] > 1, "act_bin"] = 1

        # map colors

        lut = dict(zip(df["act_bin"].unique(),

                       ["#c1f80a", "#fdaa48"]
                       ))
        row_colors1 = df["act_bin"].map(lut)
        print(lut)

    # color cell lines
    if color_col2 is not None and color_col2 == "CL":
        cmap = matplotlib.colormaps["tab10"]

        rgba = cmap(0.5)[:3]

        df["CL_bin"] = 0
        lut2 = dict(zip(df[color_col2].unique(),
                        [cmap(0.4), cmap(0.2), cmap(0.3)]  # list of colors
                        ))

        row_colors2 = df[color_col2].map(lut2)

        print(lut2)
        row_colors = pd.concat([row_colors2, row_colors1], axis=1)

        # clustermap
        g = sns.clustermap(df[plot_columns],
                           cmap="bwr",
                           center=0,
                           row_colors=row_colors,
                           vmin=-50,
                           vmax=50,
                           col_cluster=col_cluster,
                           row_cluster=row_cluster,
                           figsize=(15, 15),
                           cbar_kws={'label': 'sei pred'}
                           )

        # for legend plotting
        handles = [Patch(facecolor=lut[name]) for name in lut]
        handles2 = [Patch(facecolor=lut2[name]) for name in lut2]

        # plot CL legende
        plt.legend(handles2, lut2, title='CL',
                   bbox_to_anchor=(0.8, 1),
                   bbox_transform=plt.gcf().transFigure,
                   loc='upper left')

        # plot the activity legend
        for label in [0, 1]:
            g.ax_row_dendrogram.bar(
                0, 0, color=lut[label], label=label, linewidth=0)
            l3 = g.ax_row_dendrogram.legend(title='activity',
                                            loc='upper right', ncol=1,
                                            bbox_to_anchor=(5.9, 1.15))

    else:
        g = sns.clustermap(df[plot_columns],
                           cmap="bwr",
                           center=0,
                           vmin=-50,
                           vmax=50,
                           col_cluster=col_cluster,
                           row_cluster=row_cluster,
                           figsize=(15, 15),
                           cbar_kws={'label': 'raw sei pred'}
                           )
    return g
```

```python
def reindexDfRows(df, col_name, row_order_list):
    """re-index rows for a df column based on an ordered list"""

    # set index
    df = df.set_index(col_name)
    
    # returm sorted dataframe w/ index reset
    return df.loc[row_order_list].reset_index()  

def reindexDfCols(df, col_list, **kwargs):
    """re-order df columns by list"""
    
    # someitmes, you want to freeze some of the columns
    freeze_cols = kwargs.get("freeze_cols", None)
    
    if freeze_cols is None:
        reorderd = df[col_list]
    else:
        print("freeze", list(df.columns[:freeze_cols]))
        new_col_list = list(df.columns[:freeze_cols]) + list(col_list)
        reorderd = df[new_col_list]
    
   
    
    return reorderd
```

# per class nullomer/scaffold prediction changes 

## variance between scaffolds is greater than variance between nullomers in a scaffold for any sequence class.
- in other words, effect of nullomer insertion is very similar across nullomer sequences in scaffold,
- whereas effect on nullomer insertion across scaffolds depends heavily on the scaffold. 

```python
CLASSES = ["E6", "P", "CTCF", "E12"]

stds = {}
for SEQ_CLASS in datacols:

    test = res_subs[[SEQ_CLASS, "coor", "null.id",
                     "scaffold.id", "CL",
                     "avg_activity"]].drop_duplicates().copy()

    test_pivot = pd.pivot(test,
                          index=["scaffold.id", "CL", "coor", 'avg_activity'],
                          columns="null.id",
                          values=SEQ_CLASS
                          )

    # cluster map
    if SEQ_CLASS in CLASSES:
        out = os.path.join(RE, f"{SEQ_CLASS}.delta.scaffold.x.null.pdf")

        if os.path.exists(out) is False:

            clusterdf = reindexDfRows(test_pivot.reset_index(),
                                      "scaffold.id", endog_row_order)

            g = clusterMapCL(clusterdf.set_index("scaffold.id"),
                             clusterdf.columns[4:],
                             color_col="avg_activity", color_col2="CL",
                             row_cluster=False,

                             )
            g.fig.suptitle(SEQ_CLASS)

            # save
            plt.savefig(out, bbox_inches="tight")
            matplotlib.pyplot.close()

    # check variance between rows and columns
    test_pivot = test_pivot.reset_index()
    test_pivot = test_pivot.set_index("scaffold.id")

    # compute col-wise standard deviation
    col_std = test_pivot.iloc[:, 4:].std(axis=0).reset_index()
    # variance across the scaffolds for one null
    col_std["name"] = 'scaffold_var'
    col_std["class"] = SEQ_CLASS

    # compute row-wise standard deviation
    row_std = test_pivot.iloc[:, 4:].std(axis=1).reset_index()
    row_std["name"] = 'row_var'  # variance across nulls for one scaffold.
    row_std["class"] = SEQ_CLASS

    # add row and column standard deviation
    std = pd.concat([col_std, row_std])
    stds[SEQ_CLASS] = std

    # plot
    if SEQ_CLASS in CLASSES:
        out = os.path.join(
            RE, f"{SEQ_CLASS}.delta.scaffold.x.null.variance.pdf")
        if os.path.exists(out) is False:
            fig, ax = plt.subplots(figsize=(6, 6))
            sns.histplot(row_std[0], color="b",
                         label="null var (row-wise)",
                         common_norm=False, stat="percent"
                         )
            sns.histplot(col_std[0], color="g",
                         label="scaffold var (column-wise)",
                         common_norm=False, stat='percent'
                         )
            ax.set(title=f"{SEQ_CLASS} scaffold v. nullomer variance",
                   xlabel="standard deviation"
                   )
            ax.legend()

            # save the plot

            plt.savefig(out, bbox_inches="tight")
            matplotlib.pyplot.close()
```

```python
def gridSearchLasso(X_train, y_train, standard_scaler):
    # select alpha for lasso regression? 
    cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
    
    # try alphas - penalty weight
    lasso_alphas = np.linspace(0, 100, 10)
    grid = dict()
    grid['alpha'] = lasso_alphas

    
    # scale data?
    if standard_scaler is True:
        sc = StandardScaler()
        X_train = sc.fit_transform(X_train)
        
    # run grid search
    lasso=Lasso()
    search = GridSearchCV(lasso, grid, cv=cv).fit(X_train, y_train)

    cvres = search.cv_results_
    
    # return best scores
    print("\n\nbest params, score", search.best_params_, search.best_score_, "\n\n")
    for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
        print(np.sqrt(-mean_score), params)

gridSearchLasso(X_train, y_train, True)
```

    - because alpha is zero from gridCV
