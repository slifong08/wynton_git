```python
import os, sys
import config_readwrite as crw
import pandas as pd
```

```python
config, cfn = crw.read(os.path.join(os.path.dirname(os.getcwd()), "config.ini"))
```

```python
MPRA = config['data']["hepg2"]
BED = os.path.split(MPRA)[0] + "full.bed"
```

```python
config['data']["bed"] = BED
crw.write(config, cfn)    
```

```python
df = pd.read_csv(MPRA)

df.head()
```

```python
endog = df.loc[df["oligo"].str.contains("chr"),[ "oligo"]].copy()
endog
```

```python
endog = df.loc[df["oligo"].str.contains("chr"), ["oligo"]].copy()

endog["#chr"] = endog["oligo"].apply(lambda x: "chr" + (x.split("chr")[1]).split(":")[0])
endog["start"] = endog["oligo"].apply(lambda x: ((x.split(":")[1]).split("-")[0]))
endog["end"] = endog["oligo"].apply(lambda x: ((x.split("chr")[1]).split("-")[1]))

endog = endog[["#chr", "start", "end", "oligo"]].sort_values(by = ["#chr", "start", "end"])
```

```python
endog.drop_duplicates().to_csv(BED, sep='\t', index=False)
```
