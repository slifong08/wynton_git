```python
import os, sys
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH = "/wynton/group/ahituv/biomarin/data"
FILE ="delta_rank.csv"
DATA = os.path.join(PATH, FILE)
ARCHETYPES = "/wynton/group/ahituv/biomarin/library_2/Design/biomarin-lib2-hg38-final.x.vierstra.archetypes.cleaned.table.bed"
BED = "/wynton/group/ahituv/biomarin/library_2/Design/lib2.merged.tiles.bed"
```

```python
df = pd.read_csv(DATA)
print(df.shape)
df = df.rename(columns={"name":"tile_coor"})
df.head()
```

```python
df.loc[df["name"].str.contains("shuf")]
```

# archetype matrix

```python
arch = pd.read_csv(ARCHETYPES, sep='\t')
arch
```

# enhancer tiles

```python
tiles = pd.read_csv(BED, sep='\t', header=None, 
                    names=["tile.chr", "tile.start", "tile.end", 
                                                       "enh.chr", "enh.start",
                                                      "enh.end"])
tiles.head()
```

# coordinate ids

```python
tiles["tile.coor"] = tiles["tile.chr"] + ":" + tiles["tile.start"].map(str)+ "-"+ tiles["tile.end"].map(str)
tiles["enh.coor"] = tiles["enh.chr"] + ":" + tiles["enh.start"].map(str)+ "-"+ tiles["enh.end"].map(str)
```

# add tile info

```python
arch = pd.merge(tiles[["tile.coor","enh.coor"]], arch)
```

```python
pd.merge(arch, df, how="left")
```

```python
# make a copy. 
test = arch.copy()
```

```python
test[test.columns[1:]].drop_duplicates()
```

# max prediction value per enhancer. 
- excludes redundant motifs

```python
max_pred = test[test.columns[1:]].groupby("enh.coor")[test.columns[2:]].max().reset_index()
```
