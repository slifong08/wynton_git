```python
collection = {}
valid_chrs = ['chr1',
 'chr10',
 'chr11',
 'chr12',
 'chr13',
 'chr14',
 'chr15',
 'chr16',
 'chr17',
 'chr18',
 'chr19',
 'chr2',
 'chr20',
 'chr21',
 'chr22',
 'chr3',
 'chr4',
 'chr5',
 'chr6',
 'chr7',
 'chr8',
 'chr9',
 'chrX',
 'chrY']
```

# parse dataframe to format bed columns

```python
for i in set(df[0]):
    test=df.loc[df[0]==i].copy()
    if i not in valid_chrs:
        if i !="Background":
            test["chr"] = test[0].apply(lambda x: "chr" + x.split("chr")[1])
            test["start"] = test[1].map(int)
            test["end"] = test[2].map(int)
            test["seq_id"] = test[0].apply(lambda x: x.split("_chr")[0])
        else:
            test["chr"] = test[2]
            test["start"] = test[3].map(int)
            test["end"] = test[4].map(int)
            test["seq_id"] = '.'.join([test[0].iloc[0], test[1].iloc[0], test[2].iloc[0]])
        
    else:
        test["chr"] = test[0]
        test["start"] = test[1].map(int)
        test["end"] = test[2].map(int)
    collection[i] = test
```

## check before, after parsing shape

```python
df.shape  # before shape
```

```python
concat = pd.concat(collection.values())
concat.shape  ## after shape is the same
```

## write only bed columns

```python
concat[["chr", "start", "end", "seq_id"]].drop_duplicates().to_csv(FILE, sep='\t', header=False, index=False)
```
