```python
def readHdF5(filename, path):
    """
    read hdf5 file, dataset
    
    input
        filename (str) - name of hdf5 results
        path (str) - abs path to directory
        
    require 
        h5py
        
    method 
        1. assemble file
        2. read file
        3. get dataset

    return
        file, data
    
    """
    #1
    file = os.path.join(path, filename) 
    
    #2
    f = h5py.File(file, 'r')
    
    #3
    dset = f['data']

    return f, dset

def getIndexNames():
    
    """
    return list of 21097 index names corresponding to sei prediction features
    
    input
        none
    method 
        1. make a list for the index names
        2. open the index names file 
        3. append name to index name list
    
    return
        idxnames (list) - list of index names
    """
    
    #1
    idxnames = []
    
    #2
    IDXNAMES = "/wynton/home/ahituv/fongsl/bin/sei-framework/seq_prediction_columns.txt"
    with open(IDXNAMES, "r") as reader:
        for line in reader:
            #3
            idxnames.append(line.strip("\n"))
            
    return idxnames

def arrayToDF(dset):
    
    df = pd.DataFrame(np.vstack(dset))
    print(df.shape)
    df.columns = getIndexNames() # function get column names
    
    return df.reset_index()

def getRowNames(row_textfile, path):
    """
    return list corresponding to sample names
    """
    
     #1
    rownames = {}
    file = os.path.join(path, row_textfile) 
    
    #2
    with open(file, "r") as reader:
        for line in reader:
            
            #3
            idx, name = line.strip("\n").split("\t")
            if idx !="index":
                rownames[idx]=name
    df = pd.DataFrame(rownames.items(), columns= ["index_col", "id"])
    df["index_col"]=df["index_col"].astype(int)
    return df
```

```python
n = len(glob.glob(QUERY_F))
n
```

```python
collection = {}
for n in np.arange(len(glob.glob(QUERY_F))):
    F = f"common.{MER}mers.{NMUTS}mut.nulls.fo.ext4096.{(n+1)}_predictions.h5"
    F_LABELS = f"common.{MER}mers.{NMUTS}mut.nulls.fo.ext4096.{(n+1)}_row_labels.txt"
    f, dset = readHdF5(F,PATH)

    df = arrayToDF(dset)


    ## get row names

    names = getRowNames(F_LABELS, PATH)

    # add to dataframe
    df = pd.merge(names, df, how="left", left_on ="index_col", right_on = "index")

    # drop the index_col name (redundant)
    df = df.drop(columns=["index_col", "index"])

    # format dataframe

    df["pair_id"] = df["id"].apply(lambda x: ".".join(x.split(".")[1:]))
    df["sequence_name"] = df["pair_id"].apply(lambda x: x.split(".")[0])
    df["pos"] = df["pair_id"].apply(lambda x: x.split(".")[-1])

    # add kmer info
    df["nullomer"] = False 
    df.loc[df["id"].str.contains("null"), "nullomer" ] = True
    df.loc[df["id"].str.contains("null"), "nullomer" ] = True

    df["seq"] = df['id'].apply(lambda x: x.split(".")[0])
    df = df.loc[~df["pos"].str.contains("-")]  # remove all the extra repeats of the same kmer control data
    collection[n]= df
```

```python
# combine
df = pd.concat(collection.values())
```

```python
def getPredValues(preds, track, kmer, null):

    # get raw prediction values
    kmer_pred = preds.loc[preds.index == track][kmer].iloc[0]
    null_pred = preds.loc[preds.index == track][null].iloc[0]
    ctrl_pred = preds.loc[preds.index ==
                          track][preds.columns[:-2]].T
    ctrl_pred_mean = ctrl_pred[track].mean() 
    ctrl_pred_std = ctrl_pred[track].std()
```

```python
v = 0
for COOR in test_coor:


    # COOR = "chr1:10248183-10248383"
    TCOOR = '.'.join(COOR.split(":"))  # stupid formatting

    coordf = df.loc[(df["sequence_name"] == COOR)].copy()
    # ctsv = tsv.loc[tsv["id"] == f"{TCOOR}.fwd"].copy()
    # toggle position, as there can be multiple places where kmer/nullomer can be placed in sequence.
    # I am keeping these analyses separate for now.

    for position in set(coordf["pos"]):
        cdf = coordf.loc[coordf["pos"] == position].copy()

        # melt dataframe
        test = pd.melt(cdf, id_vars=["nullomer", "sequence_name", "pair_id", "id", "pos", 'seq'],
                       var_name="track", value_name="pred_prob")

        # constrain to CLs we want to study
        CLS = ["K562", "HepG2", " WTC11"]
        cldf = test.loc[test["track"].str.contains(
            CLS[0]) | test["track"].str.contains(CLS[1])].copy()

        # keep only datasets w/ controls
        if len(set(cldf.nullomer)) > 1:

            # kmer_id
            kmer = cldf.loc[cldf['pair_id'].str.contains(
                "kmer"), "pair_id"].iloc[0]

            # nullomer_id
            null = cldf.loc[cldf['pair_id'].str.contains(
                "null"), "pair_id"].iloc[0]

            # pivota dataframe so that rows are tracks, columns are experiments, and values are the predicted probabilityes
            preds = pd.pivot(cldf, index="track", columns="pair_id",
                             values="pred_prob")  # .reset_index()

            # compute dif between original kmer sequence and mutated ones.
            dif = preds.sub(preds[kmer], axis=0).reset_index()

            # label some columns
            dif["cl"] = dif["track"].apply(
                lambda x: x.split("|")[0].split("_")[0])
            dif["feature"] = dif["track"].apply(lambda x: x.split("|")[1])
            dif["dataset_id"] = dif["track"].apply(
                lambda x: x.split("|")[2])

            # per track
            for track in dif.track.unique():

                # get raw prediction values
                kmer_pred = preds.loc[preds.index == track][kmer].iloc[0]
                null_pred = preds.loc[preds.index == track][null].iloc[0]
                ctrl_pred = preds.loc[preds.index ==
                                      track][preds.columns[:-2]].T
                ctrl_pred_mean, ctrl_pred_std = ctrl_pred[track].mean(
                ), ctrl_pred[track].std()

                # test one track by transposing predicted values to long form

                # transpose df for prediction values of one track
                # skip first column (track) and last 3 columns (cl, feature, and dataset_id information)
                tdf = dif.loc[dif.track == track, list(dif)[1:-3]].T

                # rename column as the difference between kmer and nullomer/control values for that track
                tdf.columns = ["dif"]

                # compute z score on differences compared w/ reference sequence prediction
                tdf["z"] = stats.zscore(tdf['dif'])

                null_dif = tdf.loc[tdf.index == null,
                                   "dif"].iloc[0]
                # if nullomer z score is 2 sd away.

                if abs(tdf.loc[tdf.index == null, "z"].iloc[0]) > 2 and tdf["dif"].mean() > 0.01:
                    print(track)

                    # plot
                    fig, ax = plt.subplots(figsize=(4, 4))

                    # plot distribution of prediction differences
                    sns.histplot(tdf["dif"])

                    # plot null difference
                    ax.axvline(null_dif, ls="--", c="r")

                    # name the plot
                    ax.set(title=track + "\n" + kmer,
                           xlabel="zscore-mean diff"
                           )
                    plt.show()

                    v+=1
                    
                    if v >5:
                        break
                        
                        
```

```python
                    # write data to file
                    write_info = [
                        COOR, TCOOR,
                        kmer, null,  # pair ids
                        track,  # prediction track
                        str(kmer_pred),
                        str(null_pred),
                        str(ctrl_pred_mean),
                        str(ctrl_pred_std),
                        str(null_dif)  # nullomer difference predicted
                    ]
                    if v == 0:
                        columns = ["#coord", "tsvCoor", "kmer", "null",
                                   "track", "predKmer", "predNull",
                                   "predCtrl_mean", 'predCtrl_std',
                                   "difNullKmer"
                                   ]
                        writer.write("\t".join(columns)+"\n")

                    writer.write("\t".join(write_info)+"\n")
                    v += 1
writer.close()
```
