```python
for COOR in COORS:
    F = os.path.join(
    PATH, f"common.{MER}mers.{NMUTS}mut.nulls.fo.pam.CONCAT-10.ext4096.{COOR}_predictions.h5")
    out = os.path.join(PATH,f"HepG2.K562.Pred.{COOR}.feature.summary.tsv")

    o = pd.read_csv(out, sep='\t', header=None, 
                    names=["pair_id", "mean", "feature", "endog_dif", "rank"]).drop_duplicates()

    o.shape

    # filter for features that have at least a mean score of 0.05
    a = o.groupby("feature")["endog_dif"].mean().reset_index()

    keep_features = a.loc[a["endog_dif"].abs()>=a["endog_dif"].quantile(0.95), "feature"]


    o["type"] = o["pair_id"].apply(lambda x: str(x.split("-")[-1]))
    o["pair"] = o["pair_id"].apply(lambda x: int((x.split("_")[0])))

    # pivot
    op = o.pivot(index=["pair_id", "type", "pair"], columns = 'feature', values="endog_dif")

    # keep 95% of features w/ differences from endogenous
    op = op[keep_features].reset_index()

    op["c"] = "green"
    op.loc[op["type"]=="kmer", "c"] = "blue"
    op.loc[op["type"]=="null", "c"] = "red"


    # PCA
    X, y = op[list(op)[3:-1]],  op[["pair_id","type","pair", "c"]]

    # remove endog
    X, y = X[1:], y[1:]

    # fit PCA w/ 3 components
    pca = PCA(n_components=3)
    Xt = pca.fit_transform(X)

    print(pca.explained_variance_ratio_)

    # add row colors
    lut = dict(zip(op["type"].unique(), "brg"))
    row_colors = op["type"].map(lut)

    sns.clustermap(op.iloc[1:][list(op)[3:-1]], row_colors=row_colors,
                   center=0,
                   cmap="bwr"
                   #z_score=1
                  # row_cluster=False
                  )
    plt.show()
    break
    plt.close()
                       
```
