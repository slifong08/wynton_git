```python
import re
def find(line, regex = "TTTCGA"):
    val = line.strip('\n')
    tmp = re.search(regex, val)
    if tmp is None:
        return(False)
    else:
        return(val)
    
```

```python
obama = wiki.filter(find).compute()
obama[0:5]
```
