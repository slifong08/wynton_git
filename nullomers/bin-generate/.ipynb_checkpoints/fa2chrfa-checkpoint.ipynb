```python
import gzip
import os, sys
import chr_functions
```

```python
BUILD ="rhemac10"
chr_list = chr_functions.make_chr_list()
```

```python
PATH = "/wynton/home/ahituv/fongsl/dna/" + BUILD


FASTA = os.path.join(PATH, f"{BUILD}.fa.gz") # rheMac10 if BUILD is rhemac10
```

```python
collection_dict = dict.fromkeys(chr_list, "")

with gzip.open(FASTA, "rt") as reader:
    current_chr, seq = "", ""
    l = reader.readlines()
    last = (l)[-1]  # get last line
    
    for line in l:
       
        if ">" in line: 
            if current_chr!="":
            
                collection_dict[current_chr] = "done"  # update the dictionary
                print("done", current_chr)

                # write the file
                OUT = os.path.join(PATH, current_chr+".fa")  # file to write to      
                writer = open(OUT, "w")
                writer.write(f">{current_chr}\n")
                writer.write(seq + "\n")
                writer.close()
                print("wrote", OUT)


                seq = "" # reset the sequence variable

            NAME = (line.strip("\n")).split(">")[1]  # get chromosome name
            print(NAME)
            
            current_chr=NAME
            
        elif line == last:
            
            print("lastline")
            
            # write the last chromosome
            OUT = os.path.join(PATH, NAME+".fa")  # file to write to      
            writer = open(OUT, "w")
            writer.write(f">{NAME}\n")
            writer.write(seq + "\n")
            writer.close()
            print("wrote", OUT)
        else:
            seq+=line.strip("\n")

            
```
